### データ分析ライブラリ
- 組み込み関数・組み込み型: import文なしに利用可能な関数やデータ型のこと
    - type関数，print関数，リスト，辞書など
- ライブラリ: import文が必要な関数・データ型
    - 標準ライブラリ: 追加導入なしに利用可能なライブラリ
        - mathライブラリ，randomライブラリなど
    - 外部ライブラリ: Pythonとは独立しているライブラリで，追加導入が必要
        - NumPy, pandas，Matplotlib，seaborn，scikit-learn，Pytorchなど
- ライブラリを導入する場合，最初にimport ライブラリ名と書く
    - ライブラリに含まれる関数を使用するには，ドット(.)を打ち，関数名を続けることで使用することができる
- ライブラリの特定の関数だけを導入する場合，最初に from ライブラリ名 import 関数名と書いても使用できる
    - この場合は関数名だけで使用可能
- asを使用することで，ライブラリに別名をつけて使用できる
    - numpyライブラリはnpとして使用することが多い

```python
# ライブラリの使い方
import math # mathライブラリを使用する

# 平方根を求める
r2 = math.sqrt(2)

print(r2)
```

```python
# ライブラリの特定の関数だけを関数名だけで利用する場合
# from ライブラリ名 import 関数名 とする
from math import sqrt

# 平方根を求める
r2 = sqrt(2)

print(r2)
```

```python
# asの利用例
import numpy as np # numpyをnpとして使用する

# 最小値0，最大値2，要素数11の等差数列を求める
x = np.linspace(0, 2, 11)

print(x)

# xのすべての要素の平方根を求める
y = np.sqrt(x)

print(y)
```

```python
# 応用例: 2変数の要素の一致率を求める

# Step1
import numpy as np

# 2変数用意
y1 = np.array([1, 1, 0, 1, 0, 1, 1, 0, 1, 1])
y2 = np.array([1, 1, 0, 1, 0, 1, 1, 1, 1, 1])

# Step2 
matched = (y1 == y2) # y1とy2の要素を一致しているかどうかを調べる
print(matched)
```

```python
# 応用例: 2変数の要素の一致率を求める

# Step1
import numpy as np

# 2変数用意
y1 = np.array([1, 1, 0, 1, 0, 1, 1, 0, 1, 1])
y2 = np.array([1, 1, 0, 1, 0, 1, 1, 1, 1, 1])

# Step2 
matched = (y1 == y2) # y1とy2の要素を一致しているかどうかを調べる

# Step3
# sum関数を使う（Trueを1，Falseを0として計算）
correct = sum(matched)

# 要素数はlen関数で求まる
total = len(matched)

# 精度の計算
accuracy = correct / total

print('一致数:', correct, ' 全体数:', total, ' 一致率:', accuracy)
```

```python
# 応用例: 変数の要素を0以上1以下になるように変換する（正規化を行う）

# Step1
import numpy as np

# 変数用意
y = np.array([1, 7, 5, 2])

# 最大値(max)と最小値(min)を求める
y_max = y.max() # 最大値
y_min = y.min() # 最小値

# Step2
# 最大値と最小値から正規化する
y_normal = (y - y_min) / (y_max - y_min)

print(y)
print(y_normal)
```

### matplotlibでグラフを並べる方法
- plt.figure(figsize=(a, b)) とすると，描画領域を横aインチ，縦bインチとなる
- plt.subplot()を使用することで，一つの図の中にグラフを追加できる
- サブ領域を表す変数（慣例で変数名をaxとすることが多い）をsubplot関数で生成
    - ax = plt.subplot(A, B, C)
        - A, B, C：それぞれ一桁の正の整数
        - 図を縦にA個，横にB個に分割を行う
        - 左上から数えてC番目の領域を指定
    - ax.plot()やax.scatter()にすることで，指定した領域にグラフを描画する

```python
from sklearn.datasets import load_iris
iris = load_iris()
x, y = iris.data, iris.target
columns = iris.feature_names
print(x[:5])
```

```python
plt.figure(figsize=(15, 5))

for i in range(1, 4):
    ax = plt.subplot(1, 3, i)
    ax.scatter(x[:, 0], x[:, i], c=y, cmap='rainbow')
    ax.set_title(columns[0] + ' vs ' + columns[i])
    ax.grid()

plt.tight_layout()
plt.show()
```

```python
N = x.shape[1]
u = 5

plt.figure(figsize=(u * N, u * N))

for i in range(N):
    for j in range(N):
        k = i * 4 + j + 1
        ax = plt.subplot(N, N, k)
        ax.scatter(x[:, i], x[:, j], c=y, cmap='rainbow')
        ax.set_title(columns[i] + ' vs ' + columns[j])
        ax.grid()

plt.tight_layout()
plt.show()
```
