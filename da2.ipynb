### データ可視化
- 可視化: データを人間にわかりやすいようにグラフなどで表示すること
    - pythonでは```matplotlib```を使用することが多い
    - ```import matplotlib.pyplot as plt```を行い，```plt```として使うことがほとんど

### matplotlibの使い方
1. どのようなでデータで，何のグラフを表示するかを決める 
2. タイトルや線などの追加情報を指定する
3. ```plt.show()```を行うことで，グラフを表示する

````python
import matplotlib.pyplot as plt
plt.plot([0, 100, 200], [100, 0, 200]) # (0, 100), (100, 0), (200, 200)
plt.show()
````

### 散布図

    - あるデータと別のデータに関係性があるかを目で見て確認するとき 

- pandasとmatplotlibを連携して書く場合は

- 色に関して
    - 黒: ```'k'```
    - 赤: ```'r'```
    - 緑: ```'g'```

    - 黄: ```'y'```
    - シアン: ```'c'```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[163.6, 50.5],
              [172.6, 63.3],
              [163.7, 48.5],
              [169.9, 59.8],
              [173.9, 69.8],
              [166.2, 53.7],
              [176.7, 70.3],
              [165.4, 51.2]
              ]
# カラム
col = ['身長', '体重']

df = pd.DataFrame(table_data, columns=col)
print(df)

df.plot.scatter(x='身長', y='体重', c='b')
plt.title('身長と体重の関係')
plt.show()
````

### 散布図の調整
- 散布図を書く際に，特定のプロットだけを強調させた場合は散布図を書いた後に，以下のプログラムを書く
```python
plt.plot(x座標, y座標, c='色', marker='マーカ', markersize=サイズ)
```
- マーカーに関して
    - 丸: ```'o'```
    - バツ: ```'X'```
    - 三角: ```'^'```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[163.6, 50.5],
              [172.6, 63.3],
              [163.7, 48.5],
              [169.9, 59.8],
              [173.9, 69.8],
              [166.2, 53.7],
              [176.7, 70.3],
              [165.4, 51.2]
              ]
# カラム
col = ['身長', '体重']

df = pd.DataFrame(table_data, columns=col)
print(df)

df.plot.scatter(x='身長', y='体重', c='b')

# インデックス3番目のデータを強調する
x = df.iloc[3]['身長']
y = df.iloc[3]['体重']
plt.plot(x, y, c='r', marker='^', markersize=10)

plt.title('身長と体重の関係')
plt.show()
````

### 散布図の補助線
- 散布図に補助線を書きたい場合，以下のプログラムを書く
    - 垂直線の場合
    ```python
    plt.axvline(x=x座標, c='色', linestyle='線の種類')
    ```
    - 水平線の場合
    ```python
    plt.axhline(y=y座標, c='色', linestyle='線の種類')
    ```
- 線の種類に関して
    - 直線: ```'-'```
    - 破線: ```'--'```
    - 一点鎖線: ```'-.'```
    - 点線: ```':'```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[163.6, 50.5],
              [172.6, 63.3],
              [163.7, 48.5],
              [169.9, 59.8],
              [173.9, 69.8],
              [166.2, 53.7],
              [176.7, 70.3],
              [165.4, 51.2]
              ]
# カラム
col = ['身長', '体重']

df = pd.DataFrame(table_data, columns=col)
print(df)

df.plot.scatter(x='身長', y='体重', c='b')

# インデックス3番目のデータを強調する
x = df.iloc[3]['身長']
y = df.iloc[3]['体重']
plt.plot(x, y, c='r', marker='^', markersize=10)

# 補助線を書く
plt.axvline(x, c='r', linestyle=':')
plt.axhline(y, c='r', linestyle=':')

plt.title('身長と体重の関係')
plt.show()
````

### 相関係数
- 2種類のデータにどれくらいの関係性があるかを数値化したもの

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']

df = pd.DataFrame(table_data, columns=col)
print(df)

df.plot.scatter(x='科目A', y='科目B', c='g')
plt.title('科目Aと科目Bの関係')
plt.show()

df.plot.scatter(x='科目A', y='科目C', c='g')
plt.title('科目Aと科目Cの関係')
plt.show()
````

- 右肩上がりの場合: 横軸の値が値が増えると，縦軸の値も増える
    - これを，**正の相関**とよぶ
- 右肩下がりの場合: 横軸の値が値が増えると，縦軸の値は減る
    - これを，**負の相関**とよぶ


### 相関係数
- 関係性の強さを数値で表したもの
    - 相関が強いほど，散布図のプロットが直線上になる 
- 目安

| 相関係数      | 相関の強さ  |
|-----------|--------|
| 0.7〜1.0   | 強い正の相関 |
| 0.4〜0.7   | 正の相関   |
| 0.2〜0.4   | 弱い正の相関 |
| -0.2〜0.2  | 相関無し   |
| -0.4〜-0.2 | 弱い負の相関 |
| -0.7〜-0.4 | 負の相関   |
| -1.0〜-0.7 | 強い負の相関  |

- pandasで相関係数を求める場合
```python
df.corr()['横軸の列名']['縦軸の列名']
```

- ```df.corr()```の場合，全組み合わせを表示する（相関行列を表示する）
    - 同じ列名同士は当然相関係数は1.0となる 

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']

df = pd.DataFrame(table_data, columns=col)
print(df)

df.plot.scatter(x='科目A', y='科目B', c='g')
plt.title('科目Aと科目Bの関係')
plt.show()

df.plot.scatter(x='科目A', y='科目C', c='g')
plt.title('科目Aと科目Cの関係')
plt.show()

# 相関係数
print('科目Aと科目Bの相関係数: ', df.corr()['科目A']['科目B'])
print('科目Aと科目Cの相関係数: ', df.corr()['科目A']['科目C'])

# 相関行列
print(df.corr())
````

### 回帰直線
- 横軸のある値のとき，縦軸はどのような値になるかを予測する直線
    - 誤差が一番少なくなるように求められる直線
- pandas，matplotlib, seabornで回帰直線を求める場合
```python
sns.regplot(data=df, x='横軸の列名', y='縦軸の列名', line_kws={'color':'色'})
plt.show()
```
- 回帰直線だけでは無く，散布図もプロットしてくれる
    - 直線だけでは無く，95%の確率でこの範囲に収まってくれる領域もプロットしてくれる 
        - この範囲を**信頼区間**とよぶ

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']

df = pd.DataFrame(table_data, columns=col)
print(df)


# 回帰直線をプロットする（散布図・信頼区間も同時に出力する）
sns.regplot(data=df, x='科目A', y='科目B', line_kws={'color':'g'})
plt.title('科目Aと科目Bの関係')
plt.show()

sns.regplot(data=df, x='科目A', y='科目C', line_kws={'color':'r'})
plt.title('科目Aと科目Cの関係')
plt.show()
````

- ヒストグラムを同時に出力したい場合．```sns.jointplot()```を使用する
- 使い方
```python
sns.jointplot(data=df, x='横軸の列名', y='縦軸の列名', kind='reg', line_kws={'color':'色'})
plt.show()
```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']
df = pd.DataFrame(table_data, columns=col)

# 回帰直線をプロットする（散布図・信頼区間・ヒストグラムも同時に出力する）
sns.jointplot(data=df, x='科目A', y='科目B', kind='reg', line_kws={'color':'g'})
plt.title('科目Aと科目Bの関係')
plt.show()

sns.jointplot(data=df, x='科目A', y='科目C', kind='reg', line_kws={'color':'r'})
plt.title('科目Aと科目Cの関係')
plt.show()
````

- 相関行列を色で表示する
    - 「熱そうな色ほど大きい値」「寒そうな色ほど小さい値」 など
- pandasとmatplotlibとseabornを連携して書く場合は

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']
df = pd.DataFrame(table_data, columns=col)
sns.heatmap(df.corr())
plt.show()
````

### ヒートマップの設定
- 相関係数の値を出力したい場合，```annot=True```と書く
- 最大値を設定したい場合，```vmax=値```と書く
- 最小値を設定したい場合，```vmin=値```と書く
- 中央値で色を分けたい場合，```center=値```と書く
- まとめると
```python
sns.heatmap(df.corr(), annot=True, vmax=1, vmin=-1, center=0)
plt.show()
```

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# windowsの場合
# sns.set(style='whitegrid', font=['Meiryo'])

# macの場合
sns.set(style='whitegrid', font=['Hiragino Maru Gothic Pro'])

table_data = [[100, 94, 80],
              [85, 90, 88],
              [90, 95, 70],
              [95, 90, 62],
              [80, 85, 86],
              [80, 80, 70],
              [75, 75, 79],
              [65, 70, 65],
              [65, 60, 75],
              [60, 60, 67],
              [55, 50, 75],
              [45, 50, 68],
              [45, 48, 60],
              ]
# カラム
col = ['科目A', '科目B', '科目C']
df = pd.DataFrame(table_data, columns=col)
sns.heatmap(df.corr(), annot=True, vmax=1, vmin=-1, center=0)
plt.show()
````
