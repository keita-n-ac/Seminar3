````python
import seaborn as sns
iris = sns.load_dataset('iris')
print(iris) # printでデータを表示
````

````python
print(iris.columns) # データフレーム変数名.columnsで列名を取得できる
print(iris.dtypes)  # データフレーム変数名.dtypesで列名のデータ型を取得できる(float64は小数，objectは文字列)
print(iris.shape)   # データフレーム変数名.dtypesでデータフレームの大きさを取得できる（データ数，カラム列数）
````

### データ値の把握
- ```データフレーム変数名['列名'].describe()```と書くことで，その列名における統計値を取得できる

````python
print(iris['sepal_length'].describe())
print(iris['species'].describe())
````

### データ値の把握
- ある列で使われている値の種類を把握したい場合，```df['列名'].unique()```とする．
- これにより，speciesの値で，データを分割できる

````python
print(iris['species'].unique()) # speciesに使われている値を確認
````

- ```データフレーム変数 =  データフレーム変数[条件]```で条件を満たすデータだけを抽出できる
    - つまり， ```iris[iris['species'] == 'setosa']```とすることで，speciesがsetosaのデータだけを抽出できる
    - 抽出したデータのインデックスをリセットしたい場合，```データフレーム変数.reset_index(drop=True))```と書く

````python
data1 = iris[iris['species'] == 'setosa']     # data1はsetosaのデータのみ格納
data1 = data1.reset_index(drop=True)
print(data1)

data2 = iris[iris['species'] == 'versicolor'] # data2はversicolorのデータのみ格納
data2 = data2.reset_index(drop=True)
print(data2)

data3 = iris[iris['species'] == 'virginica']  # data3はvirginicaのデータのみ格納
data3 = data3.reset_index(drop=True)
print(data3)
````

````python
# 各データに対するsepal_lengthのヒストグラム
import matplotlib.pyplot as plt
c = 'sepal_length'
data1[c].hist(color='b', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data2[c].hist(color='r', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data3[c].hist(color='g', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
plt.xlabel(c)
plt.show()
````

````python
# 各データに対するsepal_widthのヒストグラム
import matplotlib.pyplot as plt
c = 'sepal_width'
data1[c].hist(color='b', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data2[c].hist(color='r', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data3[c].hist(color='g', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
plt.xlabel(c)
plt.show()
````

````python
# 各データに対するpetal_lengthのヒストグラム
import matplotlib.pyplot as plt
c = 'petal_length'
data1[c].hist(color='b', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data2[c].hist(color='r', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data3[c].hist(color='g', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
plt.xlabel(c)
plt.show()
````

````python
# 各データに対するpetal_widthのヒストグラム
import matplotlib.pyplot as plt
c = 'petal_width'
data1[c].hist(color='b', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data2[c].hist(color='r', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
data3[c].hist(color='g', alpha=0.5) # alphaは透過度を示す，0.0が透明，1.0が不透明
plt.xlabel(c)
plt.show()
````

- petal_widthとpetal_widthであれば，setosaとそれ以外で分割できる

````python
# 各データに対するpetal_widthの箱ひげ図
import matplotlib.pyplot as plt
import pandas as pd
c = 'petal_width'

# インデックスがそろっているため各データにおけるpetal_widthを連結できる
df = pd.concat([data1[c], data2[c], data3[c]], axis=1, ignore_index=False)
print(df)

# カラムを変更
df.columns = ['setosa', 'versicolor', 'virginica']
print(df)

sns.boxplot(data=df, width=0.5)
plt.show()
````

````python
# 各データに対するpetal_lengthの箱ひげ図
import matplotlib.pyplot as plt
import pandas as pd
c = 'petal_length'

# インデックスがそろっているため各データにおけるpetal_widthを連結できる
df = pd.concat([data1[c], data2[c], data3[c]], axis=1, ignore_index=False)
print(df)

# カラムを変更
df.columns = ['setosa', 'versicolor', 'virginica']
print(df)

sns.boxplot(data=df, width=0.5)
plt.show()
````

````python
# 各データに対するsepal_widthの箱ひげ図
import matplotlib.pyplot as plt
import pandas as pd
c = 'sepal_width'

# インデックスがそろっているため各データにおけるpetal_widthを連結できる
df = pd.concat([data1[c], data2[c], data3[c]], axis=1, ignore_index=False)
print(df)

# カラムを変更
df.columns = ['setosa', 'versicolor', 'virginica']
print(df)

sns.boxplot(data=df, width=0.5)
plt.show()
````

````python
# 各データに対するsepal_lengthの箱ひげ図
import matplotlib.pyplot as plt
import pandas as pd
c = 'sepal_length'

# インデックスがそろっているため各データにおけるpetal_widthを連結できる
df = pd.concat([data1[c], data2[c], data3[c]], axis=1, ignore_index=False)
print(df)

# カラムを変更
df.columns = ['setosa', 'versicolor', 'virginica']
print(df)

sns.boxplot(data=df, width=0.5)
plt.show()
````

- 箱ひげ図からもpetal_widthとpetal_widthであれば，setosaとそれ以外で分割できる

````python
# 各データに対するsepal_widthとsepal_lengthの散布図
import matplotlib.pyplot as plt
c1 = 'sepal_width'
c2 = 'sepal_length'

plt.scatter(data1[c1], data1[c2], color='b', alpha=0.5)
plt.scatter(data2[c1], data2[c2], color='r', alpha=0.5)
plt.scatter(data3[c1], data3[c2], color='g', alpha=0.5)

plt.xlabel(c1)
plt.ylabel(c2)
plt.show()
````

- sepal_widthとsepal_lengthの散布図から，setosaとそれ以外で分割できる境界線を書くことができることがわかる

### 3次元散布図
- 3次元（3要素）の散布図は，```from mpl_toolkits.mplot3d import Axes3D```を使用する．
```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```
- 上記のように書いて，axに対してグラフ設定を書く．

````python
# 各データに対するsepal_widthとsepal_lengthとpetal_lengthの3次元散布図
from mpl_toolkits.mplot3d import Axes3D
c1 = 'sepal_width'
c2 = 'sepal_length'
c3 = 'petal_length'

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1[c1], data1[c2], data1[c3], color='b', alpha=0.5)
ax.scatter(data2[c1], data2[c2], data2[c3], color='r', alpha=0.5)
ax.scatter(data3[c1], data3[c2], data3[c3], color='g', alpha=0.5)

# ラベルの設定はset_xlabelを使う
ax.set_xlabel(c1)
ax.set_ylabel(c2)
ax.set_zlabel(c3)

plt.show()
````

````python
# 各データに対するsepal_widthとsepal_lengthとpetal_lengthの3次元散布図
# グラフの視点を変える
from mpl_toolkits.mplot3d import Axes3D
c1 = 'sepal_width'
c2 = 'sepal_length'
c3 = 'petal_length'

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1[c1], data1[c2], data1[c3], color='b', alpha=0.5)
ax.scatter(data2[c1], data2[c2], data2[c3], color='r', alpha=0.5)
ax.scatter(data3[c1], data3[c2], data3[c3], color='g', alpha=0.5)

# ラベルの設定はset_xlabelなどを使う
ax.set_xlabel(c1)
ax.set_ylabel(c2)
ax.set_zlabel(c3)

# 視点を変える
ax.view_init(0, 240) # 縦角度，横角度

plt.show()
````

- sepal_width，sepal_length，petal_lengthの三次元散布図から，それぞれを分割できる境界線を書くことができることがわかる
